## The Marbles Game

The Challenge: You have a row of marbles of length n.  Each marble is numbered with no duplicates (zero through n-minus-1), but the marbles are not in order.

You must put the marbles in order, but you can only do so by either:

1. Switching the first two marbles, or
2. Rotating the first marble to the last position.

Submission: Create a class Marbles, with methodss for switch, rotate, and solver, such that the class and it's methods will output an ordered list.

```python
class Marbles:
    
    #Here is the function that switches the first and second position:
    def switch(self, alist):
        alist[0], alist[1] = alist[1], alist[0]
        return(alist)

    #Here is the function that moves the first position to the last:
    def rotate(self, alist):
        alist.append(alist[0])
        del alist[0]
        return(alist)
    
    #Here is the solver function:
    def solver(self, alist):
        
        # If the list is already in order:
        if alist == sorted(alist):
            print(alist)
            print("The list is already sorted.")
            exit()

        print(alist)
      
        # While loop, while the list remains unordered:
        while alist != sorted(alist):
            
            # If position 0 is one greater than position 1 (ie 2 and 1), switch the two and rotate.  
            if alist[0] - 1 == alist[1]:
                self.switch(alist)
                self.rotate(alist)
            
            # If position 0 is one less than position 1 (ie 1 and 2), rotate the first to the end.
            elif alist[0] + 1 == alist[1]:
                self.rotate(alist)
                
            # If position 0 is generically greater, rotate the two.
            elif alist[0] > alist[1]:
                self.rotate(alist)
            
            # If position 0 is generically lower, switch the two.
            else:
                self.switch(alist)
                
            print(alist)

        print("All sorted!")
```

```python
# This creates a randomized list, which is the initial input of the MarblesBoard game.
import random
listlength = int(input("Enter list length:"))
rawlist = random.sample(range(listlength), listlength)

m = Marbles()
m.solver(rawlist)
```
